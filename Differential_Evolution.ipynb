```python
import random as rd
import numpy as np
```

```python
def Fun(u):
  z = (1-u[0])**2 + 100 * (u[1]- u[0]**2)**2 + 100*(u[2]- u[1]**2)**2
  return z
```

```python
def de(n = 10, F = 0.7, Cr = 0.9):
  # Iteration Process
  tol = 10**(-5)
  N_iter = 0

  # Simple bounds
  Lb = np.array[-1,-1,-1]
  Ub = np.array[2,2,2]

  # Dimension of the search variable
  d = np.shape(Lb)

  # Initialize the population/solution

  Sol = [[]]
  Fitness = []
  for i in range(1,n):
    Sol[i] = Lb + (Ub - Lb) * np.random.random_sample(d)
    Fitness[i] = Fun(Sol(i))

  # Find the current best path
  fmin = min(Fitness)
  I = Fitness.index(fmin)
  best = Sol[I]

  # Start the iteration by differential evolution

  while fmin > tol:
    # obtain donor vectors by permutation
    k1 = rd.rand(0,n)
    k2 = rd.rand(0,n)
    k1sol = Sol[k1]
    k2sol = Sol[k2]

    # Random crossover index/matrix

    K = rd.randint(n,d)< Cr

    # DE/Rand/1 scheme

    V = Sol + F*(k1sol-k2sol)
    V = Sol * ((1-K) + V) * K

    # Evaluate new solutions

    for i in range(1,n):
      Fnew = Fun(V[i,:])

      # if the solution improves

      if Fnew <= Fitness[i]:
        Sol[i,:] = V[i,:]
        Fitness[i] = Fnew

      # Update the current best solution

      if Fnew < fmin:
        best = V[i,:]
        fmin = Fnew

    N_iter = N_iter + 1

  return N_iter



  

```
